```python
df=df.drop(['id','location'],axis=1)
df
```

```python
df['target']=df['target'].map({'Positive':1,'Negative':-1,'Neutral':0}) # mapping for assigning numerical values to Positive, Negative, Neutral
df
```

```python
tweets=df.text
tweets
```

```python
nltk.download('wordnet')
nltk.download('stopwords')
nltk.download('punkt')
```

```python
from nltk import TweetTokenizer
tk=TweetTokenizer()
tweets=tweets.apply(lambda x: tk.tokenize(x)).apply(lambda x: " ".join(x))
tweets
```

```python
# eg;
str1='Wonderfull @peacock@123!!'
str2=re.sub('[^a-zA-Z0-9]'," ",str1)
str2
```

```python
# removing special characters [^ is to take the opposite]

tweets=tweets.str.replace('[^a-zA-Z0-9]+',' ')
tweets
```

```python
# collecting words which having letters of 3 or more

from nltk.tokenize import word_tokenize
tweets=tweets.apply(lambda x: ' '.join([w for w in word_tokenize(x) if len(w)>=3]))
tweets
```

```python
from nltk.stem import SnowballStemmer
from nltk.tokenize import word_tokenize
stemmer=SnowballStemmer('english')

tweets=tweets.apply(lambda x: [stemmer.stem(i.lower()) for i in tk.tokenize(x)]).apply(lambda x: ' '.join(x))
tweets
```

```python
nltk.download('stopwords')
from nltk.corpus import stopwords
stop=stopwords.words('english')

tweets=tweets.apply(lambda x:[i for i in tk.tokenize(x) if i not in stop]).apply(lambda x: ' '.join(x))
tweets
```

```python
# vectorization

from sklearn.feature_extraction.text import TfidfVectorizer
vec=TfidfVectorizer()
train_data=vec.fit_transform(tweets)
```

```python
print(train_data)
```

```python
train_data.shape
```

```python
y=df['target'].values
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(train_data,y,test_size=0.30,random_state=42)
x_train
```

```python
x_test
```

```python
y_train
```

```python
y_test
```

```python
from sklearn.svm import SVC
from sklearn.naive_bayes import MultinomialNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier

svm_model=SVC()
nb_model=MultinomialNB()
dt_model=DecisionTreeClassifier()
rf_model=RandomForestClassifier()

lsb_model=[svm_model,nb_model,dt_model,rf_model]
```

```python
from sklearn.metrics import confusion_matrix,classification_report
for i in lsb_model:
  print(i)
  i.fit(x_train,y_train)
  y_pred=i.predict(x_test)
  print('********************************************************')
  print(classification_report(y_test,y_pred))
  print('********************************************************')
```
